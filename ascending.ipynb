```python
import logging

logging.basicConfig(
    filename="ascending.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def ascending(ls: list[int]) -> bool:
    """
    Check if a list of numbers is sorted in ascending order.
    Parameters:
        ls : A list of integers to check
    Returns:
        bool: True if the list is sorted in ascending order, otherwise false
    """
    logging.info(f"Checking if the list is in ascending order: {ls}")

    for i in range(len(ls) - 1):
        if ls[i] > ls[i + 1]:
            logging.info(f"List is not ascending. Found {ls[i]} > {ls[i+1]}.")
            return False
            
    logging.info("List is successfully verified as ascending.")
    return True


ls = [1, 2, 3, 4, 5]
result = ascending(ls)

print(result)
```
