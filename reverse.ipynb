```python
def reverse(s: str) -> str:
    """
     Reverses a given string by prepending each character.

     Args:
        s(str): The input string to be reversed.

    Returns:
        str: The fully reversed string.
    """
    ch = ''
    for i in s:
        ch = i + ch
    return ch

s = input("enter a string")
print(reverse(s))
```
