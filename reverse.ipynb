```python
import logging

logging.basicConfig(
    filename="reverse.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def reverse(s: str) -> str:
    """
    Reverses a given string by appending each character to the newly string using a for loop.
    Parameters:
        s(str): The input string to be reversed
    Returns:
        str: The fully reversed string
    """
    logging.info(f"Reversing the input string: '{s}'")
    
    ch = ''
    for i in s:
        ch = i + ch
        
    logging.info(f"String successfully reversed: '{ch}'")
    return ch


s = input("enter a string: ")
result = reverse(s)

print(result)
```
